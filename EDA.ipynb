# 히스토그램을 통한 모든 변수의 대략적인 분포 확인

```python
sns.distplot(data['유입량'], fit = norm)
(mu, sigma) = norm.fit(data['유입량'])
skewness = skew(data['유입량'])
plt.legend(['Normal dist\n$\mu=$ {:.2f} and $\sigma=$ {:.2f}\nskewness= {:.2f}'.format(mu, sigma, skewness)], loc = 'best')
plt.show()

#유입량 변수의 히스토그램 right skewness >> 정규분포와는 상당히 차이가 있어 보임
```

```python
sns.distplot(np.log(data['유입량']), fit = norm)
(mu, sigma) = norm.fit(np.log(data['유입량']))
skewness = skew(np.log(data['유입량']))
plt.legend(['Normal dist\n$\mu=$ {:.2f} and $\sigma=$ {:.2f}\nskewness= {:.2f}'.format(mu, sigma, skewness)], loc = 'best')
plt.show()

#유입량 변수에 로그변환을 한 그래프
#정규분포에 상당히 가까워짐
```

```python
#데이터집단 1, 2
from tqdm import tqdm

mu = []
sigma = []
skewness = []

fig, axes = plt.subplots(4, 4)

for i in tqdm(range(0, 4)):
  for j in range(0, 4):
    if i*4 + j + 6 <= 19:
      sns.distplot(ax = axes[i, j], a = data.iloc[:,i*4 + j + 6], fit = norm)
      mu.append(norm.fit(data.iloc[:, i*4 + j + 6])[0])
      sigma.append(norm.fit(data.iloc[:, i*4 + j + 6])[1])
      skewness.append(skew(data.iloc[:, i*4 + j + 6]))
```

```python
#데이터집단 3, 4
fig, axes = plt.subplots(4, 4)

for i in tqdm(range(0, 4)):
  for j in range(0, 4):
    if i*4 + j + 20 <= 33:
      sns.distplot(ax = axes[i, j], a = data.iloc[:,i*4 + j + 20], fit = norm)
      mu.append(norm.fit(data.iloc[:, i*4 + j + 20])[0])
      sigma.append(norm.fit(data.iloc[:, i*4 + j + 20])[1])
      skewness.append(skew(data.iloc[:, i*4 + j + 20]))
```

```python
import statsmodels.graphics.tsaplots as sgt 
from matplotlib.collections import PolyCollection

fig = plt.figure(figsize=(25,25))
for i in range(1,24):
    tmpp = data[data['홍수사상번호'] == i]
    ax1 = fig.add_subplot(5, 5, i)
    sgt.plot_pacf(tmpp['유입량'], lags = 20, zero = False, method = ('ols'), ax=ax1, color='#DC948A', vlines_kwargs={"colors":'#DC948A'})
    ax1.set_title("PACF")
    for item in ax1.collections:
        if type(item)==PolyCollection:
            item.set_facecolor('#ECC7C5')

plt.show()

plt.show()
```

# 절기

```python
def solar_term(df):   #절기 구분(7월 7일: 소서, 7월 23일: 대서)
  if (df.월 == 7 and df.일 < 7) or df.월 == 6:
    df.solar_term = 1
  elif df.월 == 7 and df.일 >= 7 and df.일 <= 23:
    df.solar_term = 2
  else:
    df.solar_term = 3

  return df
```
