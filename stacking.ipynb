# data load

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/tmp/train_std.csv')
```

```python
test1 = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/tmp/check0_std.csv')
```

```python
test2 = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/tmp/check1_std.csv')
```

```python
submit = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/tmp/submit_std.csv')
```

```python
def xy_split(df):
  y = df[['유입량']].astype(float)
  X = df.drop(['date', '홍수사상번호', '유입량', '연', '월', '일', '시간'], axis=1).astype(float)

  y = y.reset_index(drop=True)
  X = X.reset_index(drop=True)

  return y, X

y_train, X_train = xy_split(train)  #1 ~ 23번 홍수사상
y_val, X_val = xy_split(test1)   #24번 홍수사상
y_test, X_test = xy_split(test2) #25번 홍수사상
```

```python
result_submit = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/result_submit.csv')
result_test = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/result_test.csv')
result_train = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/result_train.csv')
result_val = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/result_val.csv')
```

```python
ml_submit = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/result_submit_ml_v1.csv')
ml_test = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/result_test_ml_v1.csv')
ml_train = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/result_train_ml_v1.csv')
ml_val = pd.read_csv('/content/drive/MyDrive/빅콘2021/제출용 파일/result_val_ml_v1.csv')
```

```python
import matplotlib.pyplot as plt
ax = plt.gca()
np.expm1(result_val).plot(ax=ax, legend=False)
np.expm1(y_val).plot(ax=ax, legend=False)
```

```python
ax = plt.gca()
np.expm1(result_test).plot(ax=ax, legend=False)
np.expm1(y_test).plot(ax=ax, legend=False)
```

# 평균

```python
result_1 = result_test.mean(axis=1)
```

```python
result_2 = ml_test.Label
```

```python
avg_stacking = (result_1 + result_2) /2
```

```python
ax = plt.gca()
np.expm1(avg_stacking).plot(ax=ax,legend=False,color='#DC948A')
np.expm1(y_test).plot(ax=ax,legend=False,color='#3A3C68')
```

```python
from sklearn.metrics import mean_squared_error as mse

np.sqrt(mse(np.expm1(y_test), np.expm1(avg_stacking)))
```

```python
result_v_1 = result_val.mean(axis=1)
```

```python
result_v_2 = ml_val.Label
```

```python
avg_stacking_v = (result_v_1 + result_v_2) /2
```

```python
ax = plt.gca()
np.expm1(avg_stacking_v).plot(ax=ax,legend=False,color='#DC948A')
np.expm1(y_val).plot(ax=ax,legend=False,color='#3A3C68')
```

```python
from sklearn.metrics import mean_squared_error as mse

np.sqrt(mse(np.expm1(y_val), np.expm1(avg_stacking_v)))
```

# median

# 선형회귀
